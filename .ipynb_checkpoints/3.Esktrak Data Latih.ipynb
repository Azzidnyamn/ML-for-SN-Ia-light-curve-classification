```python
import pandas as pd
```

```python
# Define the file path
file_path = 'DATA/barubarulabel.txt'

# Read the file into a DataFrame
df = pd.read_csv(file_path, delimiter=',')
df
```

```python
# Load the dataset
data = pd.read_csv('DATA/2. DATA_Pval/DATAwithPVAL_0_0.csv')
data
```

```python
dataset = pd.merge(data, df[['ZTFID', 'type']], left_on='objectId', right_on='ZTFID', how='left')
dataset
```

```python
# Replace '-' with 'unclassified' in the 'type' column
dataset['type'] = dataset['type'].replace('-', 'Candidate')
dataset['type'] = dataset['type'].fillna('Candidate')
```

```python
# Drop the redundant 'ZTFID' column if it's not needed
dataset.drop(columns=['ZTFID'], inplace=True)

# Add 'is_ia' column
dataset['is_ia'] = dataset['type'].apply(lambda x: 1 if x == 'SN Ia' else 0)
```

```python

# Reorder columns
columns = list(dataset.columns)
columns.remove('type')  # Remove 'type' from the list of columns
columns.remove('is_ia')  # Remove 'is_ia' from the list of columns
columns.insert(1, 'type')  # Insert 'type' at position 2
columns.insert(2, 'is_ia')  # Insert 'is_ia' at position 3

# Reorder the DataFrame columns based on the new column order
dataset = dataset[columns].reset_index(drop=True)

# Save the DataFrames to CSV files (optional)
# dataset.to_csv('DATA/dataset_0_0.csv', index=False)
# data_unclassified.to_csv('DATA/dataset_unclassified_0_0.csv', index=False)
```

```python
# Create two separate datasets
data_candidate = dataset[dataset['type'] == 'Candidate'].reset_index(drop=True)
dataset = dataset[dataset['type'] != 'Candidate'].reset_index(drop=True)
```
