### Ekstraks Kurva Cahaya dari fp_hists (dalam bentuk magnitudo)

```python

# Fungsi untuk mengekstrak data dari fp_hists
def extract_data(prv_candidates):
    prv_data = ast.literal_eval(prv_candidates)
    curve_g = []
    nobs_g = 0
    curve_r = []
    nobs_r = 0
    for item in prv_data:
        if item['fid'] == 1 and item['magpsf'] is not None and item['sigmapsf'] is not None and not np.isnan(item['magpsf']) and not np.isnan(item['sigmapsf']):
            if item['magpsf'] >= 0 and item['sigmapsf'] >= 0:
                curve_g.append([item['jd'], item['magpsf'], item['sigmapsf']])
                nobs_g += 1
        elif item['fid'] == 2 and item['magpsf'] is not None and item['sigmapsf'] is not None and not np.isnan(item['magpsf']) and not np.isnan(item['sigmapsf']):
            if item['magpsf'] >= 0 and item['sigmapsf'] >= 0:
                curve_r.append([item['jd'], item['magpsf'], item['sigmapsf']])
                nobs_r += 1
    return curve_g, nobs_g, curve_r, nobs_r

# Buat kolom baru dan isi dengan data yang diekstrak
df['curve_g'], df['nobs_g'], df['curve_r'], df['nobs_r'] = zip(*df['prv_candidates'].apply(extract_data))

# Tampilkan DataFrame hasil
df[['objectId', 'curve_g', 'nobs_g', 'curve_g', 'nobs_g']]
```

```python
data_curve = df[['objectId', 'curve_g', 'nobs_g', 'curve_r', 'nobs_r']]
```

### Manyatukan Data dengan ObjectId sama

```python
def parse_or_return(data):
    if isinstance(data, str):
        return ast.literal_eval(data)
    return data
    
def merge_curves(curves):
    merged = {}
    for curve in curves:
        for jd, magpsf, sigmapsf in curve:
            if jd not in merged or magpsf > merged[jd][0]:
                merged[jd] = (magpsf, sigmapsf)
    return [[jd, magpsf, sigmapsf] for jd, (magpsf, sigmapsf) in merged.items()]

def merge_duplicate_rows(group):
    objectId = group.name
    curve_g_combined = []
    curve_r_combined = []
    for _, row in group.iterrows():
        curve_g_combined.extend(parse_or_return(row['curve_g']))
        curve_r_combined.extend(parse_or_return(row['curve_r']))
    
    merged_curve_g = merge_curves([curve_g_combined])
    merged_curve_r = merge_curves([curve_r_combined])
    
    return pd.Series({
        'objectId': objectId,
        'curve_g': merged_curve_g,
        'nobs_g': len(merged_curve_g),
        'curve_r': merged_curve_r,
        'nobs_r': len(merged_curve_r)
    })
```

```python
merged_data = data_curve.groupby('objectId').apply(merge_duplicate_rows).reset_index(drop=True)
merged_data
```

### Konversi Magnitudo menjadi Fluks
#### menggunakan nilai fluks referensi 3631 Jy

```python
import numpy as np

# Fungsi untuk mengubah magnitudo ke flux dengan F0
def magnitude_to_flux(magnitudes, f0=3631):
    """
    Mengonversi magnitudo menjadi flux dengan menggunakan F0.

    Parameter:
    magnitudes (numpy array): Array magnitudo
    f0 (float): Flux referensi, default 3631 Jy

    Returns:
    numpy array: Array flux
    """
    return f0 * 10**(-0.4 * magnitudes)

# Fungsi untuk mengubah ketidakpastian magnitudo ke ketidakpastian flux dengan F0
def magnitude_uncertainty_to_flux_uncertainty(magnitudes, uncertainties, f0=3631):
    """
    Mengonversi ketidakpastian magnitudo menjadi ketidakpastian flux dengan menggunakan F0.

    Parameter:
    magnitudes (numpy array): Array magnitudo
    uncertainties (numpy array): Array ketidakpastian magnitudo
    f0 (float): Flux referensi, default 3631 Jy

    Returns:
    numpy array: Array ketidakpastian flux
    """
    fluxes = magnitude_to_flux(magnitudes, f0)
    return np.abs(0.4 * np.log(10) * fluxes * uncertainties)

# Fungsi untuk memproses curve dan menggantikan item[1] dan item[2] dengan flux dan flux_uncertainty
def process_and_replace_curve(curve, f0=3631):
    """
    Memproses curve dan menggantikan item[1] dan item[2] dengan flux dan flux_uncertainty.

    Parameter:
    curve (list of tuples): Daftar tuple dengan format (waktu, magnitudo, ketidakpastian)
    f0 (float): Flux referensi, default 3631 Jy

    Returns:
    list of tuples: Daftar tuple dengan format (waktu, flux, ketidakpastian flux)
    """
    magnitudes = np.array([item[1] for item in curve])
    uncertainties = np.array([item[2] for item in curve])
    fluxes = magnitude_to_flux(magnitudes, f0)
    flux_uncertainties = magnitude_uncertainty_to_flux_uncertainty(magnitudes, uncertainties, f0)
    new_curve = [(item[0], flux, flux_unc) for item, flux, flux_unc in zip(curve, fluxes, flux_uncertainties)]
    return new_curve
```

```python
# Mengganti item[1] dan item[2] dalam curve_g
merged_data['curve_g'] = merged_data['curve_g'].apply(process_and_replace_curve)

# Mengganti item[1] dan item[2] dalam curve_r
merged_data['curve_r'] = merged_data['curve_r'].apply(process_and_replace_curve)
merged_data
```

### Menghitung P-value dari uji statistik $\chi^2$

```python
def parse_curve_data(curve_str):
    # Check if the input is already a list
    if isinstance(curve_str, list):
        return curve_str
    # Try to safely evaluate the string
    try:
        return ast.literal_eval(curve_str)
    except (ValueError, SyntaxError):
        raise ValueError(f"Unable to parse curve data: {curve_str}")

def calculate_p_value_with_nobs(flux, flux_err, nobs):
    """
    Calculate the p-value for the given flux and flux uncertainties using chi-squared distribution,
    taking into account the number of observations (nobs).
    
    Parameters:
    flux (array-like): Array of flux values.
    flux_err (array-like): Array of flux uncertainties.
    nobs (int): Number of observations.
    
    Returns:
    float: The p-value.
    """
    # Check if there are enough observations
    if nobs <= 1:
        return np.nan
    
    # Number of observations
    n = nobs
    
    # Calculate the weighted mean of the flux values
    weights = 1 / np.square(flux_err)
    weighted_mean_flux = np.sum(flux * weights) / np.sum(weights)
    
    # Calculate the chi-squared value using the provided equation
    chi_squared = (1 / (n - 1)) * np.sum(((flux - weighted_mean_flux)**2) / np.square(flux_err))
    
    # Calculate the p-value from the chi-squared distribution with (n - 1) degrees of freedom
    p_value = chi2.sf(chi_squared * (n - 1), df=(n - 1))
    
    return p_value

def apply_p_value_calculation_with_nobs(filtered_df):
    """
    Apply p-value calculation to each row in the DataFrame for both 'g' and 'r' bands,
    using the 'nobs_g' and 'nobs_r' columns for the number of observations.
    
    Parameters:
    filtered_df (DataFrame): The input DataFrame containing 'curve_g', 'nobs_g', 'curve_r', and 'nobs_r'.
    
    Returns:
    DataFrame: The DataFrame with additional columns for p-values for 'g' and 'r' bands.
    """
    p_values_g = []
    p_values_r = []
    
    for index, row in filtered_df.iterrows():
        curve_g_str = row['curve_g']
        curve_r_str = row['curve_r']
        nobs_g = row['nobs_g']
        nobs_r = row['nobs_r']
        
        # Parse curve data
        curve_g = parse_curve_data(curve_g_str)
        curve_r = parse_curve_data(curve_r_str)
        
        # Extract flux and flux uncertainties for 'g' band
        flux_g = np.array([item[1] for item in curve_g])
        fluxunc_g = np.array([item[2] for item in curve_g])
        
        # Extract flux and flux uncertainties for 'r' band
        flux_r = np.array([item[1] for item in curve_r])
        fluxunc_r = np.array([item[2] for item in curve_r])
        
        # Calculate p-values
        p_value_g = calculate_p_value_with_nobs(flux_g, fluxunc_g, nobs_g) if len(flux_g) > 0 else np.nan
        p_value_r = calculate_p_value_with_nobs(flux_r, fluxunc_r, nobs_r) if len(flux_r) > 0 else np.nan
        
        p_values_g.append(p_value_g)
        p_values_r.append(p_value_r)
    
    # Add p-values to the DataFrame
    filtered_df['p_value_g'] = p_values_g
    filtered_df['p_value_r'] = p_values_r
    
    return filtered_df
```

```python
# Assuming merged_data is defined elsewhere and is a valid DataFrame
merged_data = apply_p_value_calculation_with_nobs(merged_data)
merged_data
```
